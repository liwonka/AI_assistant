```python
from transformers import AutoTokenizer
from auto_gptq import AutoGPTQForCausalLM


model_name = 'fffrrt/ruGPT-3.5-13B-GPTQ'
model_basename = 'gptq_model-4bit-128g'

tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
model = AutoGPTQForCausalLM.from_quantized(model_name,
        model_basename = model_basename,
        use_safetensors=True,
        trust_remote_code=True,
        device="cuda:0",
        use_triton=False,
        quantize_config=None)
```

```python
text = "Чтобы подружиться с осьминогом, нужно: \n1."

encoded_input = tokenizer(text, return_tensors='pt').to('cuda:0')
output = model.generate(
    **encoded_input,
    num_beams=4,
    max_new_tokens=160,
    no_repeat_ngram_size=2,
    # num_return_sequences=5,
    # do_sample=True
)

print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
from docx import Document
import string

def extract_clean_text(docx_filename):
    # Открываем документ .docx
    doc = Document(docx_filename)
    
    # Извлекаем текст из всех параграфов
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    
    # Соединяем параграфы в одну строку
    full_text = ' '.join(full_text)
    
    # Переводим текст в нижний регистр
    full_text = full_text.lower()
    
    # Убираем знаки препинания и табуляции
    full_text = full_text.translate(str.maketrans('', '', string.punctuation)).replace('\t', '')
    
    # Убираем лишние пробелы
    full_text = ' '.join(full_text.split())
    
    return full_text
```

```python
import docx

def create_docx_with_table(data_dict, filename):
    # Создаем новый документ
    doc = docx.Document()

    # Добавляем заголовок документа
    doc.add_heading('Таблица выполнения требований', level=1)

    # Добавляем таблицу с двумя столбцами
    table = doc.add_table(rows=1, cols=2)
    
    # Устанавливаем стиль таблицы
    table.style = 'Table Grid'

    # Добавляем заголовки
    hdr_cells = table.rows[0].cells
    hdr_cells[0].text = 'Пункт'
    hdr_cells[1].text = 'Выполнение требования'

    # Заполняем таблицу данными из словаря
    for key, value in data_dict.items():
        row_cells = table.add_row().cells
        row_cells[0].text = str(key)
        row_cells[1].text = str(value)

    # Сохраняем документ
    doc.save(filename)

# Пример словаря с пунктами и их выполнением
data = {
    'титульный лист': '+',
    'список исполнителей': '-',
    'реферат': '+',
    'содержание': '+',
    'термины и определения': '-',
    'перечень сокращений и обозначений': '-',
    'введение': '+',
    'заключение': '+',
    'список использованных источников': '+',
    'приложения': '-',
    'оценка современного состояния решаемой научно-технической проблемы': '-',
    'основание и исходные данные для разработки темы': '+',
    'обоснование необходимости проведения ВКР': '-',
    'сведения о планируемом научно-техническом уровне разработки, о патентных исследованиях и выводы из них': '-',
    'актуальность работы': '+',
    'новизна работы': '-',
    'связь данной работы с другими научно-исследовательскими работами': '-',
    'цель и задачи исследования': '+',
    'объект и предмет': '+',
    'краткие выводы по результатам работы или отдельных ее этапов': '+',
    'оценка полноты решений поставленных задач': '-',
    'разработка рекомендаций и исходных данных по конкретному использованию результатов НИР': '+',
    'результаты оценки технико-экономической эффективности внедрения': '-',
    'результаты оценки научно-технического уровня выполненной НИР в сравнении с лучшими достижениями в этой области': '-',
    'Заключение базируется на цели и задачах, выдвинутых во введении': '+',
    'Соблюдение ГОСТу': 'Неверное оформление интернет источников (8,12,16)',
    'Неверное оформление учебника (9)': 'Неверное оформление учебника (9)'
}

# Создание документа
create_docx_with_table(data, 'output_requirements_table.docx')
```
